**Razón:** Al cargar los datos y visualizarlos en un grafico se puede ver que, si se toma como variable independiente el Indice y como variable dependiente las ventas, entonces estos datos tienen una tendencia lineal en su mayor parte. Existe un par de datos que tienen un comportamiento diferente a la tendencia y, al no representar ni si quiera el 1% del dataset, seran ignorados en la seleccion del modelo, el cual es el de regresion lineal simple debido a la tendencia lineal de los datos y a la cantidad de variables que se tiene.
```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('tipo_B.csv') # Importar el dataset
x   = dataset.iloc[:, 0].values  # Variable independiente
y   = dataset.iloc[:, 1].values # Variable dependiente  
print(len(x))
print(len(y))
```

```python
## Gráfico de los datos leidos

plt.scatter(x, y, color = 'purple') 
plt.title('Datos iniciales') 
plt.xlabel('Indice') 
plt.ylabel('Ventas') 
plt.show()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)
```

```python
from sklearn.linear_model import LinearRegression

regression = LinearRegression()
regression.fit(x_train, y_train)
y_pred = regression.predict(x_test)
```

```python
## Grafico de entrenamiento

plt.scatter(x_train, y_train, color = 'green') 
plt.plot(x_train, regression.predict(x_train), color = 'purple')  
plt.title('Comparación de Indice y Ventas (Entrenamiento)') 
plt.xlabel('Indice') 
plt.ylabel('Ventas') 
plt.show()
```

```python
## Grafico de testing

plt.scatter(x_test, y_test, color = 'orange') 
plt.plot(x_test, regression.predict(x_test), color = 'darkblue') 
plt.title('Comparación de Indice y Ventas (Testing)')  
plt.xlabel('Indice') 
plt.ylabel('Ventas')  
plt.show()
```

## TIPO A

**Regresión seleccionada:**  Regresion lineal polinomial

**Razón:** Se utilizaron las variables de temperatura y ventas de helados como idependiente y dependiente respectivamente. Al graficarlas se analiza que no existe una tendencia lineal, en reaildad se tiene una forma de campana invertida (parabola). La forma que toman estos datos implica que la regresion a elegirse para este set de datos es la polinomial pues podra tomar dicha forma para poder predecir los datos en base al dataset tomado. 

```python
dataset = pd.read_csv('tipo_A.csv') # Importar el dataset
x   = dataset.iloc[:, 0].values  # Variable independiente
y   = dataset.iloc[:, 1].values # Variable dependiente  
print(len(x))
print(len(y))
```

```python
## Gráfico de los datos leidos

plt.scatter(x, y, color = 'purple') 
plt.title('Datos iniciales') 
plt.xlabel('Temp') 
plt.ylabel('Ventas') 
plt.show()
```

```python
dataset = pd.read_csv('tipo_A.csv') # Importar el dataset
x   = dataset.iloc[:, 0:1].values  # Variable independiente
y   = dataset.iloc[:, 1].values # Variable dependiente  
```

```python
from sklearn.linear_model import LinearRegression

# Crear una instancia del modelo de regresión lineal
regression = LinearRegression()

# Ajustar el modelo de regresión lineal con los datos de entrenamiento
# x_train: conjunto de datos de las variables independientes para entrenamiento
# y_train: conjunto de datos de la variable dependiente correspondiente para entrenamiento
regression.fit(x, y)
```

```python
from sklearn.preprocessing import PolynomialFeatures

poly_reg = PolynomialFeatures(degree = 3) ## A mayor grado, mejor se ajusta al modelo
## Si el modelo está muy parecido a la realidad entonces hay que dudar del sobreajuste
## En la vida real no se pasa el grado 4 incluso

x_poly  = poly_reg.fit_transform(x)

lin_reg_2 = LinearRegression()
lin_reg_2.fit(x_poly, y)
```

```python
## Gráfico de regresión lineal polinomica

# Dibujar los puntos reales del conjunto de entrenamiento
plt.scatter(x, y, color = 'green') 

# Dibujar la línea de regresión sobre el conjunto de entrenamiento
plt.plot(x, lin_reg_2.predict(x_poly), color = 'purple')  

# Establecer el título del gráfico con un texto actualizado
plt.title('Modelo de regresión polinómica') 

# Establecer la etiqueta del eje X
plt.xlabel('Temperatura') 

# Establecer la etiqueta del eje Y
plt.ylabel('Ventas $') 

# Mostrar el gráfico
plt.show()
```
