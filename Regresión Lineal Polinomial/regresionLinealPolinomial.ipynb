```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
## Lectura del archivo y selección de variables

dataset = pd.read_csv('Position_Salaries.csv') # Importar el dataset
x   = dataset.iloc[:, 1:2].values  # Variable independiente
y   = dataset.iloc[:, 2].values # Variable dependiente  
```

```python
## Regresión lineal y predicción

# Importar la clase LinearRegression de sklearn.linear_model
from sklearn.linear_model import LinearRegression

# Crear una instancia del modelo de regresión lineal
regression = LinearRegression()

# Ajustar el modelo de regresión lineal con los datos de entrenamiento
# x_train: conjunto de datos de las variables independientes para entrenamiento
# y_train: conjunto de datos de la variable dependiente correspondiente para entrenamiento
regression.fit(x, y)
```

```python
from sklearn.preprocessing import PolynomialFeatures

poly_reg = PolynomialFeatures(degree = 3) ## A mayor grado, mejor se ajusta al modelo
## Si el modelo está muy parecido a la realidad entonces hay que dudar del sobreajuste
## En la vida real no se pasa el grado 4 incluso

x_poly  = poly_reg.fit_transform(x)

lin_reg_2 = LinearRegression()
lin_reg_2.fit(x_poly, y)
```

```python
## Gráfico de regresión lineal normal

# Dibujar los puntos reales del conjunto de entrenamiento
plt.scatter(x, y, color = 'green') 

# Dibujar la línea de regresión sobre el conjunto de entrenamiento
plt.plot(x, regression.predict(x), color = 'purple')  

# Establecer el título del gráfico con un texto actualizado
plt.title('Modelo de regresión lineal') 

# Establecer la etiqueta del eje X
plt.xlabel('Cargo') 

# Establecer la etiqueta del eje Y
plt.ylabel('Sueldo $') 

# Mostrar el gráfico
plt.show()
```

```python
## Gráfico de regresión lineal polinomica

# Dibujar los puntos reales del conjunto de entrenamiento
plt.scatter(x, y, color = 'green') 

# Dibujar la línea de regresión sobre el conjunto de entrenamiento
plt.plot(x, lin_reg_2.predict(x_poly), color = 'purple')  
plt.plot(x, regression.predict(x), color = 'blue')  

# Establecer el título del gráfico con un texto actualizado
plt.title('Modelo de regresión polinómica') 

# Establecer la etiqueta del eje X
plt.xlabel('Cargo') 

# Establecer la etiqueta del eje Y
plt.ylabel('Sueldo $') 

# Mostrar el gráfico
plt.show()
```
