```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
## Lectura del archivo y selección de variables

dataset = pd.read_csv('Data.csv') # Importar el dataset
x   = dataset.iloc[:, :-1].values  # Variable independiente
y   = dataset.iloc[:, 1].values # Variable dependiente  
```

```python
## Separar el dataset para aprendizaje y testing

# Importar la función train_test_split desde la librería sklearn.model_selection
from sklearn.model_selection import train_test_split

# Dividir los datos en conjuntos de entrenamiento y prueba
# x_train, x_test: contienen las variables independientes para entrenamiento y prueba, respectivamente
# y_train, y_test: contienen las variables dependientes para entrenamiento y prueba, respectivamente
# test_size = 1/3: indica que el 33% de los datos se utilizará para el conjunto de prueba
# random_state = 0: fija la semilla para el generador de números aleatorios, asegurando la reproducibilidad
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 1/3, random_state = 0)
```

```python
## Regresión lineal y predicción

# Importar la clase LinearRegression de sklearn.linear_model
from sklearn.linear_model import LinearRegression

# Crear una instancia del modelo de regresión lineal
regression = LinearRegression()

# Ajustar el modelo de regresión lineal con los datos de entrenamiento
# x_train: conjunto de datos de las variables independientes para entrenamiento
# y_train: conjunto de datos de la variable dependiente correspondiente para entrenamiento
regression.fit(x_train, y_train)

# Utilizar el modelo entrenado para hacer predicciones sobre el conjunto de datos de prueba
# x_test: conjunto de datos de las variables independientes para prueba
# y_pred: contiene las predicciones de la variable dependiente basadas en x_test
y_pred = regression.predict(x_test)
```

```python
## Gráfico de datos de entrenamiento

# Dibujar los puntos reales del conjunto de entrenamiento
plt.scatter(x_train, y_train, color = 'green') 

# Dibujar la línea de regresión sobre el conjunto de entrenamiento
plt.plot(x_train, regression.predict(x_train), color = 'purple')  

# Establecer el título del gráfico con un texto actualizado
plt.title('Comparación de Salario y Experiencia (Entrenamiento)') 

# Establecer la etiqueta del eje X
plt.xlabel('Experiencia Laboral (años)') 

# Establecer la etiqueta del eje Y
plt.ylabel('Ingreso Anual (en $)') 

# Mostrar el gráfico
plt.show()
```

```python
## Gráfico de datos de testing

# Dibujar los puntos reales del conjunto de prueba
plt.scatter(x_test, y_test, color = 'orange') 

# Dibujar la línea de regresión utilizando el conjunto de prueba
plt.plot(x_test, regression.predict(x_test), color = 'darkblue') 

# Establecer el título del gráfico con un texto actualizado
plt.title('Comparación de Salario y Experiencia (Prueba)')  

# Establecer la etiqueta del eje X
plt.xlabel('Experiencia Laboral (años)') 

# Establecer la etiqueta del eje Y
plt.ylabel('Ingreso Anual (en $)')  

# Mostrar el gráfico
plt.show()
```
